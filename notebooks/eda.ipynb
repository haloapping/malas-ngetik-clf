# Import Modules

```python
import pandas as pd

from pandas_profiling import ProfileReport
```

# Raw Dataset

## Train Dataset

### Report

```python
train_df = pd.read_csv("../datasets/raw/train.csv", index_col="PassengerId")

train_report = ProfileReport(train_df, "Train Report", explorative=True)
train_report
```

```python
train_report.to_file("../analysis_report/raw/train_report.html")
```

### Correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax_pearson, ax_spearman, ax_kendall) = plt.subplots(nrows=1,
                                                          ncols=3,
                                                          tight_layout=True,
                                                          figsize=(15, 5),
                                                          sharey=True)

ax_pearson.set_title("Pearson Correlation")
sns.heatmap(train_df.corr("pearson"), annot=True, ax=ax_pearson)

ax_spearman.set_title("Spearman Correlation")
sns.heatmap(train_df.corr("spearman"), annot=True, ax=ax_spearman)

ax_kendall.set_title("Kendall Correlation")
sns.heatmap(train_df.corr("kendall"), annot=True, ax=ax_kendall);
```
