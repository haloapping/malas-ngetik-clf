# Report

```python
import pandas as pd

from pandas_profiling import ProfileReport
```

```python
train_df = pd.read_csv("../datasets/train_feature_engineering.csv", index_col="PassengerId")

train_report = ProfileReport(train_df, "Train Report", explorative=True)
train_report
```

```python
train_report.to_file("../html/train_report.html")
```

```python
test_df = pd.read_csv("../datasets/test_feature_engineering.csv", index_col="PassengerId")

test_report = ProfileReport(test_df, title="Test Report", explorative=True)
test_report
```

```python
test_report.to_file("../html/test_report.html")
```

# Correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax_pearson, ax_spearman, ax_kendall) = plt.subplots(nrows=1,
                                                          ncols=3,
                                                          tight_layout=True,
                                                          figsize=(15, 5),
                                                          sharey=True)

ax_pearson.set_title("Pearson Correlation")
sns.heatmap(train_df.corr("pearson"), annot=True, ax=ax_pearson)

ax_spearman.set_title("Spearman Correlation")
sns.heatmap(train_df.corr("spearman"), annot=True, ax=ax_spearman)

ax_kendall.set_title("Kendall Correlation")
sns.heatmap(train_df.corr("kendall"), annot=True, ax=ax_kendall);
```
